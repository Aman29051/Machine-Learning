It also has meadian and 1.5x interquartile range 
...

Violin plots are often used to compare the distribution of a given variable across some categories

# Heatmaps

A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors.

```python
flights=sns.load_dataset('flights')
#sns.get_dataset_names()
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
#light color more passanger 
sns.heatmap(x,cbar=False,linewidths=1,annot=True,fmt='d');  #annot=shows value #fmt='d' shows int value
```

```python
sns.heatmap(x,cbar=False,linewidths=1,annot=True,fmt='d',cmap='viridis');
```

```python
sns.heatmap(x,cbar=False,linewidths=1,annot=True,fmt='d',cmap='summer');
```

```python
sns.heatmap(x,cbar=False,linewidths=1,annot=True,fmt='d',cmap='winter');
```

# Clustermap

```python
sns.clustermap(x,row_cluster=False,annot=True,fmt='d',metric='correlation');
```

```python
sns.clustermap(x,row_cluster=False,annot=True,z_score=0,metric='correlation');
```

# Jointplot

jointplot creates a multi-panel figure that shows both bivariate(or joint) relationship between two variables along
with the univariate(or marginal) distribution of each on separate axes.

we can use jointplots in the following 4 manners:

1. Scatterplots
2. Hexbin Plots
3. KdePlots
4. Regplots

```python
sns.jointplot(x='total_bill',y='tip',data=data);
```

```python
#hex  # 2d histogram
sns.jointplot(x='total_bill',y='tip',kind='hex',data=data);
```

```python
#kde #pdf=probability distribution funtion shows higher probability
#kde=kernel density estimation
#consider black region as mountain peak

sns.jointplot(x='total_bill',y='tip',kind='kde',data=data);
```

```python
#reg
sns.jointplot(x='total_bill',y='tip',kind='reg',data=data);
```

# pairplots

Pairplots are used to plot multiple pairwise bivariate distribution in a dataset.This creates a matrix of axes and shows the relationship for each pair of columns in a dataframe. By default, it also draws the univariate distribution of each variable on the diagonal Axes.

# Distplots

Distplots are used to do univariate analysis on a continous/numerical data;


Distplot combines 3 things in one graph:
1. Histogram
2. kde(Kernel Density Estimation (PDF))
3. Rugplot

```python
titanic = sns.load_dataset('titanic')
titanic.head() #age = 177 null    #deck = 688 null   #shape = (891, 15)
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

# plotting a Distplot

```python
sns.distplot(titanic['age'],color='black',kde=False,bins=10);
```

```python
sns.distplot(titanic['age'],color='black',hist=False);
```

```python
sns.distplot(titanic['age'],color='blue',kde=False,hist=False,rug=True);
```

# How to use Distplot

```python
sns.distplot(titanic[titanic['survived']==1]['age']);
sns.distplot(titanic[titanic['survived']==0]['age']);
```
