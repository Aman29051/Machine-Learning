```python
train["Product_Category_2"].fillna(train["Product_Category_2"].median(), inplace=True)
train["Product_Category_3"].fillna(train["Product_Category_3"].median(), inplace=True)

#test

test["Product_Category_2"].fillna(train["Product_Category_2"].median(), inplace=True)
test["Product_Category_3"].fillna(train["Product_Category_3"].median(), inplace=True)
```

```python

train=train.drop('Product_ID',axis=1)
test=test.drop('Product_ID',axis=1)

for label, content in train.items():
    if pd.api.types.is_string_dtype(content):
        print(label)
```

```python
X=train.drop('Purchase',axis=1)
y=train['Purchase']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Age", "Gender", "City_Category", "Stay_In_Current_City_Years"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot, 
                                 categorical_features)],
                                remainder = "passthrough")
transformed_X = transformer.fit_transform(X)
transformed_X.shape
```

```python
pd.DataFrame(transformed_X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(transformed_X,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_jobs=-1,
                              random_state=42,
                               min_samples_split=6,
                               min_samples_leaf=7,
                               max_features=0.5,
                               n_estimators=60,
                               max_depth=None
                               )
model.fit(X_train, y_train)

# %time
# from sklearn.model_selection import RandomizedSearchCV

# # # Different RandomForestRegressor hyperparameters
# rf_grid = {"n_estimators":np.arange(10,100,10),
#             "max_depth": [None,3,5,10],
#             "min_samples_split":np.arange(2,20,2),
#             "min_samples_leaf":np.arange(1,20,2),
#             "max_features":[0.5,1,"sqrt","auto"]}

#  # Instantiate RandomizedSearchCV
# model = RandomizedSearchCV(RandomForestRegressor(n_jobs=-1,
#                                                      random_state=42),
#                                param_distributions=rf_grid,
#                                n_iter=200,
#                                cv=5,
#                                verbose=True)

#  # Fit the RandomizedSearchCV model
# model.fit(X_train,y_train)
```

```python
lin_pred = model.predict(X_test)
lin_pred
```

```python
m=mean_squared_error(y_test,lin_pred)
rmse = math.sqrt(m)
rmse
```
